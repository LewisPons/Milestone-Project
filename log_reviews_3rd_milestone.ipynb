```python
import apache_beam as beam
import re
```

```python
regex_date = r'logDate>(.*?)</logDate'
regex_device = r'device>(.*?)</device'
regex_location = r'location>(.*?)</location>'
regex_os = r'os>(.*?)</os>'
regex_ip = r'ipAddress>(.*?)</ipAddress>'
regex_phone = r'phoneNumber>(.*?)</phoneNumber>'
```

```python
with beam.Pipeline() as pipeline:
  raw_csv = (
      pipeline
      | '1. read csv' >> beam.io.ReadFromText('/content/log_reviews.csv', skip_header_lines=True)
      
  )
  
  dates = (
      raw_csv 
      | '2. apply date regex' >> beam.Regex.find(regex_date)
      | '3. slice dates' >> beam.Map(lambda x: x[9:18])
   )
  
  device = (
      raw_csv
      | '4. apply device regex' >> beam.Regex.find(regex_device)
      | '5. slice device' >> beam.Map(lambda x:  x.replace('device>',"").replace('</',''))

  )

  location = (
      raw_csv
      | '6. apply location regex' >> beam.Regex.find(regex_location)
      | '7. slice location' >> beam.Map(lambda x: x.replace('location>',"").replace('</','')) 
      #OJO CON ESTADO RARO, USE DISTINC
  )

  os = (
      raw_csv
      | '8. apply os regex' >> beam.Regex.find(regex_os)
      | '9. slice os' >> beam.Map(lambda x: x.replace('os>',"").replace('</','')) 
      #OJO CON OS RARO, USE DISTINC
  )
  
  ip = (
      raw_csv
      | '10. apply ip regex' >> beam.Regex.find(regex_ip)
      | '11. slice ip' >> beam.Map(lambda x: x.replace('ipAddress>',"").replace('</',''))
  )

  phone = (
      raw_csv
      | '12. apply phone regex' >> beam.Regex.find(regex_phone)
      | '13. slice phone' >> beam.Map(lambda x: x.replace('phoneNumber>','').replace('</',''))
      | beam.Map(print)
  )
```

```python
import re 
with beam.Pipeline() as p2:
  raw_csv = (
      p2
      | '1. read csv' >> beam.io.ReadFromText('/content/log_reviews.csv', skip_header_lines=True)
      | beam.Map(lambda x: x.split(',') )
      | beam.Map(lambda x : x + [re.search(r'logDate>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x : x + [re.search(r'phoneNumber>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x : x + [re.search(r'device>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x : x + [re.search(r'location>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x : x + [re.search(r'os>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x : x + [re.search(r'ipAddress>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x : x + [re.search(r'phoneNumber>(.*?)<', x[1]).group(1)] )
      | beam.Map(lambda x: x[0:1] + x[2:])
      | beam.Map(lambda x: ','.join(x))
      | beam.Map(print)
      | beam.io.WriteToText('/content/logReviewsClean')
      )

```
